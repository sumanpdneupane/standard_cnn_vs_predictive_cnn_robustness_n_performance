### Class Distribution

##### a. Class counts

```python
mnist_counts = class_distributions.get_class_counts(mnist_train_dataset)
fmnist_counts = class_distributions.get_class_counts(f_mnist_train_dataset)
cifar10_counts = class_distributions.get_class_counts(cifar_10_train_dataset)

datasets_info = [
    (mnist_counts, mnist_train_dataset.classes, "MNIST", "steelblue"),
    (fmnist_counts, f_mnist_train_dataset.classes, "Fashion-MNIST", "darkorange"),
    (cifar10_counts, cifar_10_train_dataset.classes, "CIFAR-10", "mediumseagreen"),
]

class_distributions.plot_all_class_distributions(datasets_info, save=False)
```

##### b. Is the dataset balanced or imbalanced?

```python
class_distributions.balance_analysis_table([
    ("MNIST", mnist_counts),
    ("Fashion-MNIST", fmnist_counts),
    ("CIFAR-10", cifar10_counts),
], save=False)
```
